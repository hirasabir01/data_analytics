```
import matplotlib.pyplot as plt

import seaborn as sns
import pandas as pd
```

```
path = "https://raw.githubusercontent.com/RamiKrispin/coronavirus/master/csv/coronavirus.csv"

df = pd.read_csv(path)
df.head()
```

```
df.info()
```

```
sns.countplot(x=df.type)
plt.show()
```

```
sns.barplot(x = "country", y = "cases", data = df)
plt.show()
```

```
sns.set(rc={"figure.figsize":(48,20)})
sns.lineplot(x = "country", y = "cases", data = df)
plt.xticks(rotation = 90)
plt.show()
```

```
sns.reset_orig()
sns.barplot(x = "country", y = "cases",hue = "type", ci = None, data = df)
plt.show()
```
